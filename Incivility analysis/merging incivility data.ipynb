In annotated_issue_level, we add a new column, consequence_category, showing the predominant consequence (positive, negative, neutral). We then make a new dataframe, basically adding to each row from annotated_comment_level, the issue trigger, target and consequence_category. This will be done merging on issue_id. This format will then enable nesting at issue_id level in STATA.

```python
annotated_issue_level.rename(columns={' consequences': 'consequences'}, inplace=True)
annotated_issue_level['consequences'] = annotated_issue_level['consequences'].apply(ast.literal_eval)
```

```python
sentiment_map = {
    'Provided technical explanation': 'positive',
    'Discontinued further discussion': 'neutral',
    'Accepting criticism': 'positive',
    'Escalating further': 'negative',
    'Trying to stop the incivility': 'neutral',
    'None': 'neutral',
    'Invoke Code of Conduct': 'negative'
}
```

```python
annotated_issue_level['positive'] = annotated_issue_level['consequences'].apply(
    lambda row: sum(sentiment_map[item] == 'positive' for item in row)
)
annotated_issue_level['negative'] = annotated_issue_level['consequences'].apply(
    lambda row: sum(sentiment_map[item] == 'negative' for item in row)
)
annotated_issue_level['neutral'] = annotated_issue_level['consequences'].apply(
    lambda row: sum(sentiment_map[item] == 'neutral' for item in row)
)
```

```python
annotated_issue_level
```

```python
# Function to determine the category
def get_consequence_category(row):
    if row['positive'] > row['negative']:
        return 'positive'
    elif row['negative'] > row['positive']:
        return 'negative'
    else:
        # Check if both are zero
        if row['positive'] == 0 and row['negative'] == 0:
            return 'neutral'
        else:
            return 'balanced'
```

```python
annotated_issue_level['consequence_category'] = annotated_issue_level.apply(get_consequence_category, axis=1)
```

```python
# Entitlement
perc_entitlement = short_comment_issue_merge.groupby('issue_id')['entitlement'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_entitlement, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'entitlement_y': 'perc_entitlement'}, inplace=True)

# Identity
perc_identity = short_comment_issue_merge.groupby('issue_id')['identity'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_identity, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'identity_y': 'perc_identity'}, inplace=True)

# Impatience
perc_impatience = short_comment_issue_merge.groupby('issue_id')['impatience'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_impatience, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'impatience_y': 'perc_impatience'}, inplace=True)

# Insulting
perc_insulting = short_comment_issue_merge.groupby('issue_id')['insulting'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_insulting, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'insulting_y': 'perc_insulting'}, inplace=True)

# None
perc_none = short_comment_issue_merge.groupby('issue_id')['none'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_none, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'none_y': 'perc_none'}, inplace=True)

# Irony
perc_irony = short_comment_issue_merge.groupby('issue_id')['irony'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_irony, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'irony_y': 'perc_irony'}, inplace=True)

# Mocking
perc_mocking = short_comment_issue_merge.groupby('issue_id')['mocking'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_mocking, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'mocking_y': 'perc_mocking'}, inplace=True)

# Threat
perc_threat = short_comment_issue_merge.groupby('issue_id')['threat'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_threat, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'threat_y': 'perc_threat'}, inplace=True)

# Vulgarity
perc_vulgarity = short_comment_issue_merge.groupby('issue_id')['vulgarity'].mean() * 100
short_comment_issue_merge = short_comment_issue_merge.merge(perc_vulgarity, on='issue_id', how='left')
short_comment_issue_merge.rename(columns={'vulgarity_y': 'perc_vulgarity'}, inplace=True)
```
